## 9.11 Тест 2-х групп (почти самостоятельное задание)

```python
df = pg.read_dataset('anova') 
```

```python
df.sample(n=5, random_state=42)
```

```python
df.groupby('Hair color')['Subject'].count()
```

```python
equal_var = pg.homoscedasticity(data=df, dv='Pain threshold', group='Hair color').at['levene', 'equal_var']

pg.homoscedasticity(data=df, dv='Pain threshold', group='Hair color')
```

```python
if equal_var:
    print('anova')
    res = pg.anova(data=df, dv='Pain threshold', between='Hair color')
else:
    print('welch_anova')
    res = pg.welch_anova(data=df, dv='Pain threshold', between='Hair color')

res
```

```python
if equal_var:
    print('anova')
    res = pg.pairwise_tukey(data=df, dv='Pain threshold', between='Hair color')
else:
    print('welch_anova')
    res = pg.pairwise_gameshowell(data=df, dv='Pain threshold', between='Hair color')

res
```
