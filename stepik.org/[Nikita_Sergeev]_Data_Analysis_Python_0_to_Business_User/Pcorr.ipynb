## 10.7 Корреляция для повторных замеров

- Subject номер пациента/испытуемого (у каждого человека было несколько замеров -  и важно что у каждого было РАЗНОЕ количество замеров: например, у 1 пациента было 4 замера, а у 7 только 3 замера, а у 8 было 8 замеров....)
- pH измеренное значение кислотности крови (pH)
- PaCO2 измеренное у того же пациента в тот же момент парциальное давление углекислого газа в крови в кПа

```python
df = pg.read_dataset('rm_corr')
```

```python
df.head(3)
```

```python
pg.rm_corr(data=df, x='pH', y='PacO2', subject='Subject')
```

```python
PH_LOW, PH_HIGH = 7.35, 7.45
PCO2_LOW, PCO2_HIGH = 4.7, 6.0

df['Диагноз']='Неоднозначно: метаболический или смешанный'

df.loc[(df['pH'] < PH_LOW) & (df['PacO2'] > PCO2_HIGH), 'Диагноз'] = 'Ацидоз'
df.loc[(df['pH'] > PH_HIGH) & (df["PacO2"] < PCO2_LOW), "Диагноз"] = 'Алкалоз'
df.loc[(df['pH'].between(PH_LOW, PH_HIGH)) & (df['PacO2'].between(PCO2_LOW, PCO2_HIGH)), 'Диагноз'] = 'Норма'
```
