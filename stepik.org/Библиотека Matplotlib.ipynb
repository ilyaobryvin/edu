```python
# Графики построились, но кое-где произошли наложения данных друг на друга. Для устранения этой проблемы применяют функцию plt.tight_layout().
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.title('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.title('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'y')
plt.title('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'c')
plt.title('y4')

plt.tight_layout()

plt.show()
```

### Axes.tick_params()

* labelrotation - отвечает за варьирование угла наклона меток осей координат;
* axis - позволяет повернуть метки только одной оси;
* labelcolor - меняет цвет меток осей координат;
* labelsize - позволяет менять размер шрифта меток координатных осей;
* pad - позволяет варьировать расстоянием между метками оси и границей самой координатной оси;
* direction - направление рисок на координатной оси;
* color, length, width - параметры рисок;
* colors - меняет цвет одновременно и меток координатных осей, и рисок;

#### AutoMinorLocator

* set_minor_locator
* set_minor_formatter
* ax.tick_params(which='minor')
* ax.grid(which='both')

major - (По умолчанию) все изменения, которые мы до этого производили, касались основных меток координатных осей;   
minor - производить определённые манипуляции следует с промежуточными метками осей координат;   
both - использовать для обоих меток осей координат;   

```python
from matplotlib.ticker import AutoMinorLocator
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

ax.tick_params(which='major', axis='x', pad=10)
ax.tick_params(which='minor', labelsize=7, grid_alpha=0.2)
ax.xaxis.set_minor_formatter('{x:.1f}')
ax.grid(which='both')

plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 30)
y = 2 * np.sin(x)

fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, c='royalblue', linewidth=3, marker='o') 

ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

ax.tick_params(which='major', pad=15, axis='x')
ax.tick_params(which='major', pad=10, axis='y')
ax.tick_params(which='minor', labelsize=7, grid_alpha=0.2, labelrotation=90, axis='x')
ax.xaxis.set_minor_formatter('{x:.1f}')
ax.tick_params(which='minor', labelsize=7, grid_alpha=0.2, axis='y')
ax.yaxis.set_minor_formatter('{x:.2f}')
ax.grid(which='both')

plt.show()
```

### scatter

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
df['target'] = iris_dataset['target']
df['target'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

display(df)
```

```python
x = df['petal length (cm)']
y = df['petal width (cm)']

plt.scatter(x, y)
# plt.scatter(x='petal length (cm)', y='petal width (cm)', data=df)

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target')

for name, group in groups:
    plt.scatter(group[x], group[y], label=name)

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

#### Изменение отображения точек на графике

```python
x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target')

for name, group in groups:
    plt.scatter(group[x], group[y], label=name, edgecolor='black', linewidth=0.5, alpha=0.3)

plt.grid(alpha=0.5)
plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

#### Параметры c и cmap

```python
x = df['petal length (cm)']
y = df['petal width (cm)']
z = df['petal width (cm)'] / df['petal length (cm)']

plt.scatter(x, y, c=z, cmap='summer')
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

cbar = plt.colorbar()
cbar.set_label('Отношение ширины лепестка к его длине')

plt.show()
```
