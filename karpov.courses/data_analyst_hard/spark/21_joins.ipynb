```python
#Оба DF имеют одинаковое название для колонки user_id, поэтому можно упростить предыдущий код
#Данный пример аналогичен предыдущему
df_users. \
    join(df_course_enrolments, 'user_id'). \
    show(5)
```

```python
# Выбираем все колонки из users, и course_id, course_enrolment_id из course_enrolments
df_users. \
    join(df_course_enrolments, df_users.user_id == df_course_enrolments.user_id). \
    select(df_users['*'], df_course_enrolments['course_id'], df_course_enrolments['course_enrolment_id']). \
    show()
```

```python
#Добавляем groupBy

df_users. \
    join(df_course_enrolments, df_users.user_id == df_course_enrolments.user_id). \
    groupBy(df_users['user_id']). \
    count(). \
    show()
```

```python
# Можно использовать alias для обращения к DF
df_users.alias('u'). \
    join(df_course_enrolments.alias('ce'), 'user_id'). \
    groupBy('u.user_id'). \
    count(). \
    show()
```

```python
#Left join

df_users. \
    join(df_course_enrolments, df_users.user_id == df_course_enrolments.user_id, 'left'). \
    show()
```

```python
#Так тоже работает, потому что совпадают имена колонок user_id в обоих DF
df_users. \
    join(df_course_enrolments, 'user_id', 'left'). \
    show()
```

```python
# Находим всех пользователей, кто не выбрал курс
#Также используем alias для обращения к колонкам DF
 
df_users.alias('u'). \
    join(df_course_enrolments.alias('ce'), df_users.user_id == df_course_enrolments.user_id, 'left'). \
    filter('ce.course_enrolment_id IS NULL'). \
    select('u.*', 'ce.course_id', 'ce.course_enrolment_id'). \
    show()
```

```python
#Находим для каждого студента число курсов, которые он посещает
#Если он не посещает ничего, должны получить ноль для такого студента
#Также помним, что мы можем использовать SQL для этого выполните createOrReplaceTempView Для каждого DF

from pyspark.sql.functions import sum, when

df_users.alias('u'). \
    join(df_course_enrolments.alias('ce'), df_users.user_id == df_course_enrolments.user_id, 'left'). \
    groupBy('u.user_id'). \
    agg(sum(when(df_course_enrolments['course_enrolment_id'].isNull(), 0).otherwise(1)).alias('course_count')). \
    orderBy('u.user_id'). \
    show()
```

```python
#Right Join

df_course_enrolments. \
    join(df_users, df_users.user_id == df_course_enrolments.user_id, 'right'). \
    show()
```

```python
# Находим всех пользователей, кто не выбрал курс
#Также используем alias для обращения к колонкам DF
 
df_course_enrolments.alias('ce'). \
    join(df_users.alias('u'), df_users.user_id == df_course_enrolments.user_id, 'right'). \
    filter('ce.course_enrolment_id IS NULL'). \
    select('u.*', 'course_id', 'course_enrolment_id'). \
    show()
```

```python
#Находим для каждого студента число курсов, которые он посещает
#Если он не посещает ничего, должны получить ноль для такого студента

df_course_enrolments.alias('ce'). \
    join(df_users.alias('u'), df_users.user_id == df_course_enrolments.user_id, 'right'). \
    groupBy('u.user_id'). \
    agg(sum(when(df_course_enrolments['course_enrolment_id'].isNull(), 0).otherwise(1)).alias('course_count')). \
    orderBy('u.user_id'). \
    show()
```
