```python
from pyspark.sql import Row

df_courses = spark.createDataFrame([Row(**course) for course in courses])

df_courses.show()
```

```python
#Сортируем по двум колонкам
df_courses. \
    sort('suitable_for', 'enrollment'). \
    show()
```

```python
#Сортируем по двум колонкам в разном порядке
from pyspark.sql.functions import desc

df_courses. \
    sort('suitable_for', desc('enrollment')). \
    show()
```

```python
#Пример с добавлением новой колонки,сортировкой
#Обратите внимание как добавлена новая колонка. Мы задали её отдельно сначала как course_level

from pyspark.sql.functions import col, when

course_level = when(col('suitable_for') == 'Beginner', 0).otherwise(when(col('suitable_for') == 'Intermediate', 1).otherwise(2))

df_courses. \
    withColumn('course_level', course_level). \
    orderBy(course_level, col('number_of_ratings').desc()). \
    show()
```

```python
#Также можно использовать expr и SQL
from pyspark.sql.functions import expr
 
course_level = expr("""
    CASE 
        WHEN suitable_for = 'Beginner'
            THEN 0
        WHEN suitable_for = 'Intermediate'
            THEN 1
        ELSE 2
    END
""")
 
df_courses. \
    withColumn('course_level', course_level). \
    sort(course_level, col('number_of_ratings').desc()). \
    show()
```
