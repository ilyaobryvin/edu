```python
help(df_users.drop)
```

```python
#Удалить колонку можно многими способами. Ниже три разных способа с помощью drop
from pyspark.sql.functions import col

df_users.drop('last_updated_ts').printSchema()

df_users.drop(df_users['last_updated_ts']).printSchema()

df_users.drop(col('last_updated_ts')).printSchema()
```

```python
#Если указать несуществующую колонку, то команда по сути будет проигнорирована. Будет возвращен DF без изменений.
df_users.drop(col('not_a_col')).printSchema()
```

```python
#Можно удалить несколько колонок за один раз
df_users.drop('first_name', 'last_name').printSchema()

df_users.drop('first_name', 'last_name', 'email').show()
```

```python
#Если указать несуществующую колонку, она будет проигнорирована, удаление остальных отработает корректно
df_users.drop('not_a_col', 'first_name', 'last_name', 'email').show()
```

```python
#Также можно передать список колонок для удаления
list_drop_columns = ['first_name', 'last_name', 'email', 'phone_numbers', 'not_a_col']

df_users_droped_cols = df_users.drop(*list_drop_columns)
df_users_droped_cols.show()
```
