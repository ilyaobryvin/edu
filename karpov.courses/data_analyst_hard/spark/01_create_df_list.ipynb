```python
#Spark connection
import os
import socket
from pyspark.sql import SparkSession
 
APACHE_MASTER_IP = socket.gethostbyname("apache-spark-master-0.apache-spark-headless.apache-spark.svc.cluster.local")
APACHE_MASTER_URL = f"spark://{APACHE_MASTER_IP}:7077"
POD_IP = os.environ["MY_POD_IP"]
SPARK_APP_NAME = f"spark-{os.environ['HOSTNAME']}"
JARS = "/nfs/env/lib/python3.8/site-packages/pyspark/jars/clickhouse-native-jdbc-shaded-2.6.5.jar"
MEM = "512m"
CORES = 1
 
spark = SparkSession.\
        builder.\
        appName(SPARK_APP_NAME).\
        master(APACHE_MASTER_URL).\
        config("spark.executor.memory", MEM).\
        config("spark.jars", JARS).\
        config("spark.executor.cores", CORES).\
        config("spark.driver.host", POD_IP).\
        getOrCreate()
```

```python
ages_list = [21, 23, 18, 41, 32]
type(ages_list)
```

```python
#Так можно получить помощь по методам Spark
spark
help(spark.createDataFrame)
```

```python
#Создаем DataFrame из List
df_list = spark.createDataFrame(ages_list, 'int')
spark.createDataFrame(ages_list, 'int').show()
type(df_list)
df_list.show(2)
```

```python
#Далее рассматриваем различные методы создания DF
from pyspark.sql.types import IntegerType
spark.createDataFrame(ages_list, IntegerType())
spark.createDataFrame(ages_list, IntegerType()).show()
```

```python
names_list = ['Scott', 'Donald', 'Mickey']
spark.createDataFrame(names_list, 'string')
spark.createDataFrame(names_list, 'string').show()
```

```python
from pyspark.sql.types import StringType
spark.createDataFrame(names_list, StringType())
spark.createDataFrame(names_list, StringType()).show()
```

```python
ages_list = [(21, ), (23, ), (41, ), (32, )]
type(ages_list)
```

```python
spark.createDataFrame(ages_list, 'age int')
spark.createDataFrame(ages_list, 'age int').show()
```

```python
#Создаем DF из List of Tuples
#Обратите внимания как задаем имена колонок и типы
users_list = [(1, 'Scott'), (2, 'Donald'), (3, 'Mickey'), (4, 'Elvis')]
spark.createDataFrame(users_list)
df = spark.createDataFrame(users_list, 'id int, name string')
df.show()
df.printSchema()
```

```python
#Преобразуем List of Dicts в DataFrame
users_list = [
    {'user_id': 1, 'user_first_name': 'Scott'},
    {'user_id': 2, 'user_first_name': 'Donald'},
    {'user_id': 3, 'user_first_name': 'Mickey'},
    {'user_id': 4, 'user_first_name': 'Elvis'}
]

df = spark.createDataFrame(users_list)
df.show()
df.printSchema()
```
