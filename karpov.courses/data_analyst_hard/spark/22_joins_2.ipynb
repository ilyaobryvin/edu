```python
#Пример с full join и coalesce
#Дополняем данные из df_users1 данными из df_users2, если в первом DF отсутствуют данные
 
from pyspark.sql.functions import coalesce

df_users1. \
    join(df_users2, df_users1.email == df_users2.email, 'full'). \
    select(
        coalesce(df_users1['email'], df_users2['email']).alias('email'),
        coalesce(df_users1['first_name'], df_users2['first_name']).alias('first_name'),
        coalesce(df_users1['last_name'], df_users2['last_name']).alias('last_name'),
        coalesce(df_users1['gender'], df_users2['gender']).alias('gender'),
        coalesce(df_users1['ip_address'], df_users2['ip_address']).alias('ip_address'),
    ). \
    show()
```

```python
#Broadcast Join
#Также известен как Map-Side join
#Меньший по объему DF будет реплицирован на все экзекьюторы
#Данный вид Join`а может происходить автоматически, 
#если размер DF меньше размера, задаваемого параметром spark.sql.autoBroadcastJoinThreshold
#По умолчанию 10 Mb

spark.conf.get('spark.sql.autoBroadcastJoinThreshold')
```

```python
#Также broadcast Join можно задать самостоятельно
#Представим, что df_users1 - это бОльший по объему DF
#df_users2 представим мЕньшим по объему DF
#В этом случае применение broadcast join будет выглядеть так
#Важный момент - broadcast работает как hint, то есть Spark может не применить broadcast join и выбрать другой тип
#https://spark.apache.org/docs/latest/sql-performance-tuning.html#join-strategy-hints-for-sql-queries

from pyspark.sql.functions import broadcast

df_users1.join(broadcast(df_users2), df_users1.email == df_users2.email, 'inner').show()
```

```python
#Иногда необходимо получить перекрестное соединение (декартово произведение). 
#Каждую строку первого DF соединяем с каждой строкой второго DF

df_users1. \
    crossJoin(df_users2). \
    count()
```

```python
#Если не указать условие также получим декартово произведение
df_users1. \
    join(df_users2). \
    count()
```
