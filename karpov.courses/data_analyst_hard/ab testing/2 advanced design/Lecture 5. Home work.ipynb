```python
print('Доля значений параметра в полном наборе данных:')
parameter_list = ['vas', 'platform']
for step in parameter_list:
    population_sample_parameter(df_data2, step)
```

```python
# сформируем выборку из 10 тыс. пользователей, исходя из условия,
# что параметры представлены в соотношении, как для полного набора данных
def get_strat(df, cols, n=10000):
    # df['combined'] = list(zip(df['vas'], df['platform']))
    df['combined'] = df.loc[:, cols].apply(lambda row: tuple(row[c] for c in cols), axis=1)

    #формируем столбец с комбинацией целевых параметров
    
    weight = df['combined'].value_counts(normalize=True)
    counts = df['combined'].value_counts()
    
    df['combined_weight'] = df['combined'].apply(lambda x: weight[x]/counts[x]) # рассчитываем вес каждой комбинации, по которому будет сэмплироваться выборка
    strat_sample_dataset = df.sample(n, weights=df['combined_weight']) # сэмплируем по весам комбинации параметров
    return strat_sample_dataset

get_strat(df_data2, parameter_list)
```

```python
# выведем долю значений для каждого параметра в выборке
print('Доля значений параметра в выборке из 10 тыс. пользователей:')
for step in parameter_list:
    population_sample_parameter(
        get_strat(df_data2, parameter_list), step
    )
```
